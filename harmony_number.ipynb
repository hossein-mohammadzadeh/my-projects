```python
# These codes are to observe the effect of the number of harmonics on fourier seris
import numpy as np
import matplotlib.pyplot as plot
harmony_number=int(input('Enter harmony_number: '))
y=0
time= np.arange(0,30,0.01)

for index in range(1,harmony_number,2):
  base=(1/index)*np.sin(index*time)
  y=y+base


plot.plot(time,y,linewidth=2,color='k')
plot.xlabel('time')
plot.ylabel('amplitue=sin(time)')
plot.grid(True)
plot.axhline(y=0,linewidth=2, color='k')
plot.axvline(x=0,linewidth=2, color='k')
```
